```python
dirs_i_want = []
for root, dirs, files in os.walk("./data/train_data/"):
        for file in files:
            if file.endswith('jpeg'):
                dirs_i_want.append(
                    {'number': re.findall(r'\d+', file)[0],
                     'path' : root+'/'+file,
                     'filename' : file[:-5],
                     'dir': root.split('/')[-1],
                     'type' : file.split('_')[1].split('.')[0]})

data_df = pd.DataFrame(dirs_i_want)
```
